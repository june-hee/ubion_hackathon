```python
train_data = df['제목+본문'].dropna(how = 'any') # Null 값이 존재하는 행 제거
train_data.duplicated()
print(train_data.isnull().values.any()) # Null 값이 존재하는지 확인
train_data
```

```python
from konlpy.tag import Mecab
import re
mecab = Mecab()

t_w = mecab.nouns(df['제목+본문'].to_string())
t_w
```

```python
from collections import Counter

count = Counter(t_w)
tag_count = []
tags = []

for n, c in count.most_common(100):
    dics = {'tag': n, 'count': c}
    if len(dics['tag']) >= 2 and len(tags) <= 49:
        tag_count.append(dics)
        tags.append(dics['tag'])

for tag in tag_count:
    print(" {:<14}".format(tag['tag']), end='\t')
    print("{}".format(tag['count']))
```

```python
train_data = train_data.str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
korean_stopwors_path = "/Users/ijunhui/Downloads/ubion/해커톤/korean_stopword.txt"

# 불용어 텍스트 파일 오픈
with open (korean_stopwors_path, encoding= 'utf8') as f : 
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]
```

```python

# 형태소 분석기 OKT를 사용한 토큰화 작업 (다소 시간 소요)
okt = Okt()

tokenized_data = []
for sentence in (train_data):
    tokenized_sentence = okt.morphs(sentence, stem=True) # 토큰화
    
    t_s = []
    for i in range(len(tokenized_sentence)):
        if len(tokenized_sentence[i]) > 1:
            t_s.append(tokenized_sentence[i])
        
    stopwords_removed_sentence = [word for word in t_s if not word in stopwords] # 불용어 제거
    tokenized_data.append(stopwords_removed_sentence)
```

```python
print('리뷰의 최대 길이 :',max(len(review) for review in tokenized_data))
print('리뷰의 평균 길이 :',sum(map(len, tokenized_data))/len(tokenized_data))
plt.hist([len(review) for review in tokenized_data], bins=50)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
plt.show()
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences = tokenized_data, vector_size = 100, window = 5, min_count = 1, workers = 1, sg = 0)
```
